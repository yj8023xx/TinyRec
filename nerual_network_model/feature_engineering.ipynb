````python
column_names = list(train_data.columns.values)
column_names.remove('label')
column_names
````

````python
# 数值型特征
numerical_columns = ['timestamp', 'releaseYear', 'movieRatingCount', 'movieAvgRating',
                     'movieRatingStddev', 'userRatingCount', 'userAvgReleaseYear',
                     'userReleaseYearStddev', 'userAvgRating', 'userRatingStddev']
# 类别型特征
categorical_columns = ['movieId', 'userId', 'movieGenre1', 'movieGenre2', 'movieGenre3', 'userRatedMovie1', 
                       'userRatedMovie2', 'userRatedMovie3', 'userRatedMovie4', 'userRatedMovie5', 'userGenre1', 
                       'userGenre2', 'userGenre3', 'userGenre4', 'userGenre5']
# 用户特征
user_columns = ['userId', 'timestamp', 'userRatedMovie1', 'userRatedMovie2', 'userRatedMovie3',
                'userRatedMovie4', 'userRatedMovie5', 'userRatingCount', 'userAvgReleaseYear',
                'userReleaseYearStddev', 'userAvgRating', 'userRatingStddev',
                'userGenre1', 'userGenre2', 'userGenre3', 'userGenre4', 'userGenre5']
# 物品特征
item_columns = ['movieId', 'releaseYear', 'movieGenre1', 'movieGenre2', 'movieGenre3',
                'movieRatingCount', 'movieAvgRating', 'movieRatingStddev']

# 风格特征
genre_columns = ['movieGenre1', 'movieGenre2', 'movieGenre3',
                 'userGenre1', 'userGenre2', 'userGenre3', 'userGenre4', 'userGenre5']

# 好评电影ID特征
rated_columns = ['userRatedMovie1', 'userRatedMovie2',
                  'userRatedMovie3', 'userRatedMovie4', 'userRatedMovie5']
````

````python
# genre features vocabulary 电影类型特征
genre_vocab = [
    'Film-Noir', 'Action', 'Adventure', 'Horror', 'Romance', 'War', 'Comedy',
    'Western', 'Documentary', 'Sci-Fi', 'Drama', 'Thriller', 'Crime',
    'Fantasy', 'Animation', 'IMAX', 'Mystery', 'Children', 'Musical'
]
````

## 缺失值填充

````python
train_data.isnull().sum() / len(train_data)
````

````python
test_data.isnull().sum() / len(test_data)
````

````python
for feature in genre_columns:
    train_data[feature].fillna(train_data[feature].mode()[0], inplace=True)
    test_data[feature].fillna(test_data[feature].mode()[0], inplace=True)
````

````python
for feature in rated_columns:
    train_data[feature].fillna(train_data[feature].mean(), inplace=True)
    test_data[feature].fillna(test_data[feature].mean(), inplace=True)
````

## 归一化

````python
numerical_df = train_data[numerical_columns]
numerical_df = (numerical_df - numerical_df.min()) / (numerical_df.max() - numerical_df.min())
train_data[numerical_columns] = numerical_df
````

````python
numerical_df = test_data[numerical_columns]
numerical_df = (numerical_df - numerical_df.min()) / (numerical_df.max() - numerical_df.min())
test_data[numerical_columns] = numerical_df
````

## 特征编码

````python
for feature in categorical_columns:
    feature_mapping = {val: i for i, val in enumerate(list(pd.concat([train_data[feature], test_data[feature]]).unique()))}
    train_data[feature] = train_data[feature].map(feature_mapping)
    test_data[feature] = test_data[feature].map(feature_mapping)
````

## 特征字典

生成数据集描述字典，格式如下

```json
# data/taobao_tiny_h5/feature_map.json

            "source": "",
            "type": "categorical",
            "vocab_size": 100,
            "index": 1
        },
        ...
    }
}
```

````python
feature_map = OrderedDict()
feature_map['dataset_id'] = 'movielens'
feature_map['num_fields'] = len(column_names)
````

````python
feature_map['feature_specs'] = OrderedDict()
for feature in column_names:
    feature_map['feature_specs'][feature] = {
        'source': 'user' if feature in user_columns else 'item',
        'type': 'numerical' if feature in numerical_columns else 'categorical',
        'vocab_size': 1 if feature in numerical_columns else pd.concat([train_data[feature], test_data[feature]]).nunique(),
        'index': train_data.columns.get_loc(feature)
    }
````

````python
train_data.to_csv('../data/movielens/data_for_train.csv', index=False)
test_data.to_csv('../data/movielens/data_for_test.csv', index=False)
with open('../data/movielens/feature_map.json', 'w') as obj:
    json.dump(feature_map, obj)
````
